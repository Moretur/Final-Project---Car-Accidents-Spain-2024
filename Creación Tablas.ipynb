Por último con la tabla accidentes:

```python
map_mes = {
    1: "Enero", 2: "Febrero", 3: "Marzo", 4: "Abril",
    5: "Mayo", 6: "Junio", 7: "Julio", 8: "Agosto",
    9: "Septiembre", 10: "Octubre", 11: "Noviembre", 12: "Diciembre"
}

map_dia = {
    1: "Lunes", 2: "Martes", 3: "Miercoles", 4: "Jueves",
    5: "Viernes", 6: "Sabado", 7: "Domingo"
}

def franja_horaria(h):
    if pd.isna(h):
        return "Sin dato"
    h = int(h)
    if 0 <= h < 6: return "Madrugada"
    if 6 <= h < 12: return "Mañana"
    if 12 <= h < 18: return "Tarde"
    return "Noche"

def gravedad(row):
    if row["TOTAL_MU30DF"] > 0:
        return "Mortal"
    if row["TOTAL_HG30DF"] > 0:
        return "Grave"
    if row["TOTAL_HL30DF"] > 0:
        return "Leve"
    return "Sin víctimas"
```

```python
if "MES" in accidentes_24.columns:
   accidentes_24["MES_DESC"] = accidentes_24["MES"].map(map_mes)

if "DIA_SEMANA" in accidentes_24.columns:
    accidentes_24["DIA_SEMANA_DESC"] =accidentes_24["DIA_SEMANA"].map(map_dia)

if "HORA" in accidentes_24.columns:
    accidentes_24["FRANJA_HORARIA"] = accidentes_24["HORA"].apply(franja_horaria)

accidentes_24["GRAVEDAD_DESC"] = accidentes_24.apply(gravedad, axis=1)

columnas_accidentes_exp = [
    "ID_ACCIDENTE",
    "ANYO",
    "MES", "MES_DESC",
    "DIA_SEMANA", "DIA_SEMANA_DESC",
    "HORA", "FRANJA_HORARIA",
    "TOTAL_MU30DF", "TOTAL_HG30DF", "TOTAL_HL30DF",
    "GRAVEDAD_DESC"
]

accidentes = accidentes_24[[c for c in columnas_accidentes_exp if c in accidentes_24.columns]]
```
